```python
# 1. BRONZE TABLE (Raw YouTube Data)
spark.sql("""
    CREATE TABLE IF NOT EXISTS courseify.default.course_bronze (
        course_id STRING,
        video_id STRING,
        video_title STRING,
        channel_name STRING,
        playlist_id STRING,
        ingested_at TIMESTAMP
    ) USING DELTA
""")

# 2. SILVER TABLE (AI Processed Data with Notes & QA)
spark.sql("""
    CREATE TABLE IF NOT EXISTS courseify.default.course_silver (
        course_id STRING,
        main_topic_id INT COMMENT 'Maps to syllabus 0-24',
        main_topic_name STRING COMMENT 'Syllabus Topic Name',
        sub_videos_json STRING COMMENT 'List of videos mapped to this topic',
        notes_status INT COMMENT '1 if Notes generated else 0',
        qa_status INT COMMENT '1 if QA generated else 0',
        notes_data STRING COMMENT 'Markdown notes content',
        qa_data STRING COMMENT 'JSON formatted QA data',
        updated_at TIMESTAMP
    ) USING DELTA
""")
print("✅ Bronze & Silver Tables Created Successfully!")
```
