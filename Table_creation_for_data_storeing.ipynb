```python
# Databricks Python Notebook lo SQL run cheyadaniki
spark.sql("""
    CREATE TABLE IF NOT EXISTS courseify.default.courseify_data (
        id STRING,
        title STRING COMMENT "Course Title",
        topic_id STRING COMMENT "Topic ID",
        topic STRING COMMENT "Topic Name",
        video_links STRING COMMENT "Youtube Video Links",
        channel_name String COMMENT "Youtube Channel Name",
        description STRING COMMENT "Description of the Class used for questions & notes preperation",
        notes BOOLEAN COMMENT "Is notes Generated",
        notes_data STRING COMMENT "Notes Data",
        qa BOOLEAN COMMENT "Is QA Generated",
        qa_data STRING COMMENT "questions & answers"
    ) USING DELTA
""")

print("Table successfully created!")
```
