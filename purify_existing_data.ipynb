```python
import json
import base64

def purify_existing_data():
    # Load the data you successfully generated so far
    try:
        with open("raw_data.json", "r", encoding="utf-8") as f:
            data = json.load(f)
    except FileNotFoundError:
        print("Please save your generated data as 'raw_data.json' first.")
        return

    purified_data = {
        "course_notes": {
            "py_notes_shared": {}
        }
    }

    # Loop through the topics that successfully generated
    for topic_id, content in data.get("course_notes", {}).get("py_notes_shared", {}).items():
        print(f"Purifying Topic ID: {topic_id}")
        
        # 1. Base64 Encode the Markdown Notes
        raw_notes = content.get("notes", "")
        # Encode to bytes, then base64, then back to a string
        notes_b64 = base64.b64encode(raw_notes.encode('utf-8')).decode('utf-8')
        
        # 2. Reformat the Practice Test
        purified_quiz = []
        for q_data in content.get("practice_test", []):
            purified_quiz.append({
                "question": q_data.get("q", ""),
                "options": q_data.get("opts", []),
                "answer_index": q_data.get("ans", 0)
                # Dropped 'exp' as it wasn't in your target format
            })
            
        # 3. Map to Target Structure
        purified_data["course_notes"]["py_notes_shared"][topic_id] = {
            "notes_base64": notes_b64,
            "practice_test": purified_quiz
        }

    # Save the beautifully purified data
    with open("final_courseify_data.json", "w", encoding="utf-8") as f:
        json.dump(purified_data, f, indent=4)
        
    print("\n✅ Purification Complete! Check 'final_courseify_data.json'")

if __name__ == "__main__":
    purify_existing_data()
```
